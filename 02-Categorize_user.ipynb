```python
import pandas as pd
from scipy import stats
```

```python
# Load the data to check its structure and contents
file_path = 'user_info_001_to_750_v5.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
data.info()
```

```python
# =====================================
# Step 1: Ensure 'comment_date' is in datetime format
# =====================================
# Automatically parse dates in ISO 8601 format
data['comment_date'] = pd.to_datetime(data['comment_date'], errors='coerce')

# =====================================
# Step 2: Calculate days per comment for each user (Updated Criteria 1)
# =====================================
def calculate_days_per_comment(group):
    max_date = group['comment_date'].max()
    min_date = group['comment_date'].min()
    total_comments = len(group)
    
    # Avoid division by zero for users with only one comment
    days_active = (max_date - min_date).days + 1  # Add 1 to count the first day
    if total_comments > 0:
        return days_active / total_comments
    else:
        return days_active  # If only one comment, days per comment = days active

# Apply the function to calculate days per comment
days_per_comment = data.groupby('member_name').apply(calculate_days_per_comment)

# Add the days per comment back to the original DataFrame
data['days_per_comment'] = data['member_name'].map(days_per_comment)

# Update Criteria 1: Users with days per comment <= median
median_days_per_comment = data['days_per_comment'].median()
criteria_1_users = days_per_comment[days_per_comment <= median_days_per_comment].index

# =====================================
# Step 3: Filter users with only 1 comment across the entire dataset (Criteria 2)
# =====================================
comment_counts_total = data.groupby('member_name').size()
criteria_2_users = comment_counts_total[comment_counts_total == 1].index

# =====================================
# Step 4: Calculate average ratings for each user (Updated Criteria 3)
# =====================================
rating_columns = ['taste_rating', 'env_rating', 'service_rating', 'hygiene_rating', 'fair_price_rating']
data['average_rating'] = data[rating_columns].mean(axis=1)

# Update Criteria 3: Users with average_rating >= median
median_average_rating = data['average_rating'].median()
criteria_3_users = data.groupby('member_name')['average_rating'].mean()
criteria_3_users = criteria_3_users[criteria_3_users >= median_average_rating].index

# =====================================
# Step 5: Calculate the average comment photo count per user (Updated Criteria 4)
# =====================================
average_photo_count = data.groupby('member_name')['comment_photo_count'].mean()

# Update Criteria 4: Users with average_photo_count >= median
median_photo_count = average_photo_count.median()
criteria_4_users = average_photo_count[average_photo_count >= median_photo_count].index

# =====================================
# Step 6: Segregate users into two categories
# =====================================
# suspicious_1: Meets Criteria 1, 3, and 4
category_1_users = set(criteria_1_users) & set(criteria_3_users) & set(criteria_4_users)

# suspicious_2: Meets Criteria 2, 3, and 4
category_2_users = set(criteria_2_users) & set(criteria_3_users) & set(criteria_4_users)

# Count the number of users in each category
category_1_count = len(category_1_users)
category_2_count = len(category_2_users)

# non-suspicious: Those who don't belong to suspicious_1 or suspicious_2
other_users_count = len(data['member_name'].unique()) - category_1_count - category_2_count

# =====================================
# Step 7: Update the existing "user_category" column in the DataFrame
# =====================================
def assign_user_category(member_name):
    if member_name in category_1_users:
        return "suspicious_1"
    elif member_name in category_2_users:
        return "suspicious_2"
    else:
        return "non-suspicious"

# Update the existing "user_category" column with the new categories
data['user_category'] = data['member_name'].apply(assign_user_category)

# =====================================
# Step 8: Display the results
# =====================================
print("Category Counts:")
print(f"Suspicious 1 Users: {category_1_count}")
print(f"Suspicious 2 Users: {category_2_count}")
print(f"Non-Suspicious Users: {other_users_count}")

# Display the updated DataFrame with the updated "user_category" column and new "days_per_comment" column
print(data[['member_name', 'user_category', 'days_per_comment']].drop_duplicates())
print(data)

# Print the median of member days per comment, num of comment, average rating, and member photo count
print("Median Days Per Comment:", median_days_per_comment)
print("Median Number of Comments:", comment_counts_total.median())
print("Median Average Rating:", median_average_rating)
print("Median Member Photo Count:", median_photo_count)
```
